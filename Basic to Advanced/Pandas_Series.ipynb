# Pandas Series
A pandas series is like a column in a table. it is a 1-D array holding data or any type.

**Series from lists**

```python
import numpy as np
import pandas as pd
```

```python
# string
country = ['india', 'pakistan', 'china']
pd.Series(country)
```

```python
# integer
runs = [100, 200, 300]
pd.Series(runs)
```

```python
# custom index
marks = [67,57,89,99]
subjects = ['maths', 'science', 'english', 'hindi']
pd.Series(marks, index=subjects)
```

```python
# setting a name
marks = pd.Series(marks, index=subjects, name='Raja ke marks')
marks
```

# Series from dict

```python
marks = {
    'maths': 67,
    'science': 57,
    'english': 89,
    'hindi': 99
}
marks_series = pd.Series(marks)
marks_series
```

**Series Attributes**

```python
# size
marks_series.size
```

```python
# dtype
marks_series.dtype
```

```python
# name
marks_series.name
```

```python
# is_unique
marks_series.is_unique
```

```python
# index
marks_series.index
```

```python
# values
marks_series.values
```

**Series Indexing**

```python
# ineger indexing
x = pd.Series([12,34,56,78,1,34,54,23,32,99])
x[0]
```

**Series with python Functionalites**

```python
# len / type / dir / sorted / max / min
print(len(x))
print(type(x))
print(dir(x))
print(sorted(x))
print(max(x))
print(min(x))
```

```python
# type conversion
print(list(x))
print(tuple(x))
print(set(x))
print(dict(x))
```

```python
x
```

```python
# membership operator
print(12 in x)
print(12 not in x)
print(12 in x.values)
print(12 not in x.values)
```

```python
# loops
for i in x:
  print(i)
```

```python
# arithmetic operations
print(x + 10)
print(x - 10)
print(x * 10)
print(x / 10)
```

```python
# Relational operators
print(x > 50)
print(x < 50)
print(x >= 50)
print(x <= 50)
print(x == 50)
print(x != 50)
```
